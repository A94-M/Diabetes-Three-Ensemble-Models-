```python
x=data.drop('Outcome',axis=1)
y=data['Outcome']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30)
```

# create_model

```python
model1=LogisticRegression()
model2=GradientBoostingClassifier()
model3=RandomForestClassifier()
```

```python
result=[]
```

```python
def calling(model):
    model.fit(x_train,y_train)
    pre=model.predict(x_test)
    accur=accuracy_score(pre,y_test)
    result.append(accur)
    print(model)
    print('accuracy is :',accur)
```

```python
calling(model1)
```

```python
calling(model2)
```

```python
calling(model3)
```

```python
models=['LogisticRegression','GradientBoostingClassifier','RandomForestClassifier']
KL=pd.DataFrame({'Models':models,
             "accuracy_scor":result})
```

```python
plt.plot(KL.Models ,KL.accuracy_scor,marker='o',linestyle='dashdot')
```
